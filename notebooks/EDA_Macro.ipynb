```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Visualization 1

```python
df_gdp = pd.DataFrame({'Year': range(2000, 2023), 'GDP_Growth': np.random.normal(3, 2, 23)})
px.line(df_gdp, x='Year', y='GDP_Growth', title='Global GDP Growth').show()
```

### Visualization 2

```python
df_inf = pd.DataFrame({'Year': range(2000, 2023), 'Inflation': np.random.normal(2, 1.5, 23)})
px.bar(df_inf, x='Year', y='Inflation', title='Global Inflation').show()
```

### Visualization 3

```python
df_yield = pd.DataFrame({'Maturity': [1,2,3,5,7,10,20,30], 'Yield': [4.1, 4.0, 3.9, 3.8, 3.9, 4.1, 4.4, 4.5]})
px.line(df_yield, x='Maturity', y='Yield', title='Yield Curve').show()
```

### Visualization 4

```python
df_fx = pd.DataFrame({'Day': range(100), 'FX_Rate': np.random.normal(0, 1, 100).cumsum()})
px.line(df_fx, x='Day', y='FX_Rate', title='FX Exchange Rate Volatility').show()
```
